```python
import pandas as pd
import json
import glob
import os
```

```python
def join_files():
    all_files = glob.glob(os.path.join(r'./Data/', "*.xlsx"))

    df = pd.read_excel(all_files[0])
    
    column_list = ['Country', 'Year', 'Rank', 'Total', 'C1: Security Apparatus',
       'C2: Factionalized Elites', 'C3: Group Grievance', 'E1: Economy',
       'E2: Economic Inequality', 'E3: Human Flight and Brain Drain',
       'P1: State Legitimacy', 'P2: Public Services', 'P3: Human Rights',
       'S1: Demographic Pressures', 'S2: Refugees and IDPs',
       'X1: External Intervention']
    
    df = df[column_list]

    for a in range(1, len(all_files)):
        frame = pd.read_excel(all_files[a])
        frame = frame[column_list]
        df = pd.concat([df, frame])

    return df
```

```python
df = join_files()
df.head()
```

```python
df['Country'] = df['Country'].apply(lambda x: x.strip())
```

```python
dc = pd.read_csv('compare.csv', encoding='latin_1')
dc.head()
```

```python
df = df.drop(columns=['Country'])
df.head()
```

```python
df = df.rename(columns={'Preferred': 'country', 'continent': 'continent', 'C1: Security Apparatus': 'c1',
       'C2: Factionalized Elites': 'c2', 'C3: Group Grievance': 'c3', 'E1: Economy': 'e1',
       'E2: Economic Inequality': 'e2', 'E3: Human Flight and Brain Drain': 'e3',
       'P1: State Legitimacy': 'p1', 'P2: Public Services': 'p2', 'P3: Human Rights': 'p3',
       'S1: Demographic Pressures': 's1', 'S2: Refugees and IDPs': 's2',
       'X1: External Intervention': 'x1', 'Year': 'year', 'Total': 'total'})

df
```

```python
indicators = {'c1': 'Cohesion', 'c2': 'Cohesion', 'c3': 'Cohesion', 'e1': 'Economic', 'e2': 'Economic', 
              'e3': 'Economic', 'p1': 'Political', 'p2': 'Political', 'p3': 'Political', 's1': 'Social', 
              's2': 'Social', 'x1': 'Social', 'total': 'Total'}
```

```python
indicator_names = {'c1': 'Security Apparatus', 'c2': 'Factionalized Elites', 'c3': 'Group Grievance', 
                  'e1': 'Economy', 'e2': 'Economic Inequality', 'e3': 'Human Flight and Brain Drain', 
                  'p1': 'State Legitimacy', 'p2': 'Public Services', 'p3': 'Human Rights', 
                  's1': 'Demographic Pressures', 's2': 'Refugees and IDPs', 'x1': 'External Intervention', 
                  'total': 'Total'}
```

```python
indicator_categories = {'c1': 'cat1', 'c2': 'cat2', 'c3': 'cat3', 'e1': 'cat1', 'e2': 'cat2', 
              'e3': 'cat3', 'p1': 'cat1', 'p2': 'cat2', 'p3': 'cat3', 's1': 'cat1', 
              's2': 'cat2', 'x1': 'cat3', 'total': 'Total'}
```

```python
df = pd.melt(df, id_vars=['country', 'iso3', 'continent', 'year', 'Rank'])
df.head()
```

```python
df['indicator'] = df['variable'].map(indicators)
df['indicator_name'] = df['variable'].map(indicator_names)
df['indicator_cat'] = df['variable'].map(indicator_categories)
df
```

```python
df['value'] = df['value'].round(2)
```

```python
df['year'].drop_duplicates().sort_values()
```

```python
df = df.rename(columns={'year': 'date'})
```

```python
df['year'] = df['date'].dt.year
df
```

```python
df['rank'] = df['Rank'].str.extract('(\d+)').astype('Int64')
df
```

```python
df['rank_continent'] = df.groupby(['year', 'variable', 'continent'])['rank'].rank('min', ascending=True)
df
```

```python
world_rank = df.groupby(['variable', 'year'], as_index=False)['rank'].count().rename(columns={'rank': 'country_count'})
world_rank = world_rank[['year', 'country_count']].drop_duplicates()
world_rank
```

```python
year_list = world_rank['year'].drop_duplicates().values.tolist()
year_dict = {}

for y in year_list:
    year_dict[y] = world_rank.loc[world_rank['year']==y, 'country_count'].iloc[0]
#     year_dict[str(y)] = world_rank[(world_rank['year']==y)]['country_count'].values[0].astype(int)

    
year_dict
```

```python
with open('fsiWorldMaxData.js', 'w') as f:
    f.write('var fsiWorldMaxData = ')
    f.write(json.loads(json.dumps(str(year_dict))))
    f.close()
```

```python
continent_rank = df.groupby(['variable', 'year', 'continent'], as_index=False)['rank'].count().rename(columns={'rank': 'country_count'})
continent_rank = continent_rank[['continent', 'year', 'country_count']].drop_duplicates()
continent_rank
```

```python
continent_list = continent_rank['continent'].drop_duplicates().values.tolist()
continent_dict = {}

year_list = continent_rank['year'].drop_duplicates().values.tolist()

for c in continent_list:
    year_dict = {}
    for y in year_list:
        year_dict[y] = continent_rank[(continent_rank['year']==y) & (continent_rank['continent']==c)]['country_count'].values[0]
    continent_dict[c] = year_dict
    
continent_dict
```

```python
with open('fsiContinentMaxData.js', 'w') as f:
    f.write('var fsiContinentMaxData = ')
    f.write(json.loads(json.dumps(str(continent_dict))))
    f.close()
```

```python
df_diff = df[df['variable']=='total'][['iso3', 'value', 'year']]
df_diff = df_diff.pivot(index='iso3', columns='year', values='value')
df_diff = df_diff.diff(axis=1)
df_diff
```

```python
df_diff = df_diff.reset_index(drop=False).melt(id_vars='iso3')
df_diff.head()
```

```python
df_diff['value'] = df_diff['value'].round(3)
df_diff.head()
```

```python
with open('fsiDiffData.js', 'w') as f:
    f.write('var fsiDiffData = ')
    f.write(json.loads(json.dumps(df_diff.sort_values(by='year').groupby('iso3')['value'].apply(list).to_json(orient='index'))))
    f.close()
```

```python
temp = df[['indicator', 'indicator_name', 'indicator_cat']].drop_duplicates().reset_index(drop=True)
temp
```

```python
iso_list = df['iso3'].drop_duplicates().sort_values().values.tolist()
iso_dict = {}

ind_list = df['indicator'].drop_duplicates().values.tolist()
for iso in iso_list:
    ind_dict = {}
    for ind in ind_list:
        tp = temp[temp['indicator']==ind].reset_index()
        t_dict = {}
        for t in range(len(tp)):
            
            val = df[(df['indicator']==ind) & (df['indicator_name']==tp['indicator_name'][t]) & (df['iso3']==iso)] \
            .sort_values(by='year').groupby('iso3')['value'].apply(list)[0]
            
            t_dict[tp['indicator_cat'][t]] = {'category_name': tp['indicator_name'][t], 'value': val}
            
        ind_dict[ind] = t_dict
    iso_dict[iso] = ind_dict
    
iso_dict
```

```python
with open('fsiCountryData.js', 'w') as f:
    f.write('var fsiCountryData = ')
    f.write(json.loads(json.dumps(str(iso_dict))))
    f.close()
```

```python
total_df = df[(df['indicator']=='Total')].sort_values(by='year').groupby('iso3')['value'].apply(list).reset_index()
total_dict = {}

for t in range(len(total_df)):
    total_dict[total_df['iso3'][t]] = total_df['value'][t]
    
total_dict
```
